```python
import requests

# Define your OpenCage API key
api_key = "YOUR_API_KEY"

# Define the base URL for the API
base_url = "https://api.opencagedata.com/geocode/v1/json"

# Load your data with locations
#locations = [
   # {"latitude": 12.9716, "longitude": 77.5946},
    #{"latitude": 19.0760, "longitude": 72.8777},
    # Add more locations here
#]

# Initialize a list to store geocoded results
geocoded_results = []

# Batch process locations
for location in df:
    lat = df["Lattitude"]
    lon = df["Longitude"]

    # Prepare the query parameters
    params = {
        "q": f"{lat},{lon}",
        "key": api_key,
    }

    # Make the API request
    response = requests.get(base_url, params=params)
    data = response.json()

    # Process the response and extract relevant information (e.g., city, state)
    if "results" in data and data["results"]:
        result = data["results"][0]
        city = result.get("components", {}).get("city", "")
        state = result.get("components", {}).get("state", "")

        geocoded_results.append({"latitude": lat, "longitude": lon, "city": city, "state": state})

# Print or save the geocoded results
for result in geocoded_results:
    print(f"Latitude: {result['latitude']}, Longitude: {result['longitude']}, City: {result['city']}, State: {result['state']}")
```
